```python
df = pd.read_csv('data.csv')
df
```

```python
# removing empty (NaN) values from the df
df.dropna(axis=0)
df
```

```python
df.fillna({'Calories': 300})
```

```python
# finding statistical values: mean, mode, median
mode_cal = df["Calories"].mode()
mode_cal

mean_dur = df["Duration"].mean()
mean_dur

mean_cal = df["Calories"].mean()
print('mean calories:', mean_cal)

df.fillna({'Calories': mean_cal})
```

```python
# fixing format problems
# convering non datetime format to datetime one

df["Date"] = pd.to_datetime(df["Date"], format = 'mixed')
df.dropna(subset=['Date'])
```

```python
# Changing the value of an individual cell

df.loc[7, 'Duration'] = 45
df
```

```python
# dropping a whole column
df.drop(columns=['Maxpulse'])
```

```python
# identifying duplicate values

df.duplicated().head(15)
```

```python
prices = np.random.randint(10, 25, size=(4, 3))
df = pd.DataFrame(prices, columns=list('abc'), index=['Apple', 'Graips', 'Pear', 'Cherry'])
df
```

```python
df.name = 'Fruit prices'
df.index.name = 'Fruit'
df.columns.name = 'Price'
df
```

```python
# using function to and in df
df.apply(np.sum)
```

```python
df = np.transpose(df)
df
```

```python
# inner function to df by apply() method
df.apply(lambda x: x.max() - x.min(), axis=0)
```

```python
df.columns.name = ''
```

```python
def dif(x):
  return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
df.apply(dif)
```

```python
# applying a function to df elements by map() method
df.map(lambda x: 2*x)
```

#Ranking

```python
# rank method
df_cars = pd.read_csv('/content/CAR DETAILS FROM CAR DEKHO.csv')
df_cars
```

```python
df_cars.sort_values(by='selling_price', ascending=False).head(10)
```

```python
df_cars['rank'] = df_cars['km_driven'].rank(method= 'average')
df_cars.sort_values('rank', ascending=False).head(10)
```
